```python
import numpy as np
#import easyocr
import pandas as pd
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "True"   # MKL 충돌 잠시 우회
import easyocr  
import time
```

```python

reader = easyocr.Reader(['ko', 'en'], gpu=True)
```

```python
import os                      # 표준 라이브러리
#from PIL import Image          # 예: 이미지 미리 확인용
#import easyocr                 # 예: OCR

# 1️⃣ 이미지가 들어 있는 폴더 경로
img_dir = "neuro_patient_images"     # ← 자신의 폴더명으로 수정

# 2️⃣ 디렉터리 내 모든 *.png 경로 수집 (정렬 포함)
image_paths = sorted(
    os.path.join(img_dir, fname)
    for fname in os.listdir(img_dir)
    if fname.lower().endswith(".jpg")           # 확장자 필터
)

print(f"총 {len(image_paths)}장 불러옴")
```

```python
len(image_paths)
```

```python
#ocr_results = []
#image_paths = [f"data_example/neuro_patient_images{i+1}.png" for i in range(30)]
import time
# 각 이미지에 대해 OCR 수행
t0 = time.perf_counter()

ocr_results = []
for path in image_paths:
    result = reader.readtext(path)
    for _, text, conf in result:
        ocr_results.append({
            "filename": path.split('/')[-1],
            "text": text,
            "confidence": round(conf, 2)
        })

df_ocr = pd.DataFrame(ocr_results)

# ── 2. 시간 측정 종료 ──────────────────────────────────────
elapsed = time.perf_counter() - t0
```
