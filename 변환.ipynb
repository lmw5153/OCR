```python
# -*- coding: utf-8 -*-
import gzip
import csv

def phred33_to_q(qual_str):
    """ASCII  Phred   """
    return [ord(c) - 33 for c in qual_str]

infile = "ERR4833012_1.fastq.gz"
outfile = "sample.csv"

with gzip.open(infile, "rt") as f, open(outfile, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["read_id", "sequence", "quality_scores"])  # 
    
    while True:
        try:
            read_id = f.readline().strip()
            if not read_id:  # 
                break
            seq = f.readline().strip()
            f.readline()  # +  
            qual = f.readline().strip()
            
            q_scores = " ".join(map(str, phred33_to_q(qual)))
            writer.writerow([read_id, seq, q_scores])
        except Exception as e:
            print("Error:", e)
            break

print(" :", outfile)
```

```python
# -*- coding: utf-8 -*-
#    FASTQ.GZ  CSV 
import gzip
import csv
from pathlib import Path

def phred33_to_q(qual_str):
    """ASCII  Phred   """
    return [ord(c) - 33 for c in qual_str]

#     (: "./data")
input_dir = Path("raw_data/HV")   #  
files = list(input_dir.glob("*.fastq.gz"))  # fastq.gz   

for infile in files:
    outfile = infile.with_suffix("").with_suffix(".csv")  
    # : ERR4833012_1.fastq.gz -> ERR4833012_1.csv

    with gzip.open(infile, "rt") as f, open(outfile, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["read_id", "sequence", "quality_scores"])  # 

        while True:
            try:
                read_id = f.readline().strip()
                if not read_id:  # 
                    break
                seq = f.readline().strip()
                f.readline()  # +  
                qual = f.readline().strip()

                q_scores = " ".join(map(str, phred33_to_q(qual)))
                writer.writerow([read_id, seq, q_scores])
            except Exception as e:
                print("Error:", e)
                break

    print(" :", outfile)
```

```python
import zipfile
from pathlib import Path

# CSV    
input_dir = Path("raw_data/HV")   #  
output_zip = input_dir / "HV.zip"   #   

# CSV  
csv_files = list(input_dir.glob("*.csv"))

# 
with zipfile.ZipFile(output_zip, "w", compression=zipfile.ZIP_DEFLATED) as zipf:
    for csv_file in csv_files:
        zipf.write(csv_file, arcname=csv_file.name)  # arcname    
        print(":", csv_file.name)

print(" :", output_zip)
```
