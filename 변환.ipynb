```python
# -*- coding: utf-8 -*-
import gzip
import csv

def phred33_to_q(qual_str):
    """ASCII → Phred 품질 점수 리스트"""
    return [ord(c) - 33 for c in qual_str]

infile = "ERR4833012_1.fastq.gz"
outfile = "sample.csv"

with gzip.open(infile, "rt") as f, open(outfile, "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["read_id", "sequence", "quality_scores"])  # 헤더
    
    while True:
        try:
            read_id = f.readline().strip()
            if not read_id:  # 끝
                break
            seq = f.readline().strip()
            f.readline()  # + 줄 스킵
            qual = f.readline().strip()
            
            q_scores = " ".join(map(str, phred33_to_q(qual)))
            writer.writerow([read_id, seq, q_scores])
        except Exception as e:
            print("Error:", e)
            break

print("저장 완료:", outfile)
```

```python
# -*- coding: utf-8 -*-
# 폴더 내 모든 FASTQ.GZ 파일을 CSV로 변환
import gzip
import csv
from pathlib import Path

def phred33_to_q(qual_str):
    """ASCII → Phred 품질 점수 리스트"""
    return [ord(c) - 33 for c in qual_str]

# 변환할 폴더 경로 지정 (예: "./data")
input_dir = Path("raw_data/HV")   # 현재 폴더
files = list(input_dir.glob("*.fastq.gz"))  # fastq.gz 파일 모두 찾기

for infile in files:
    outfile = infile.with_suffix("").with_suffix(".csv")  
    # 예: ERR4833012_1.fastq.gz -> ERR4833012_1.csv

    with gzip.open(infile, "rt") as f, open(outfile, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["read_id", "sequence", "quality_scores"])  # 헤더

        while True:
            try:
                read_id = f.readline().strip()
                if not read_id:  # 끝
                    break
                seq = f.readline().strip()
                f.readline()  # + 줄 스킵
                qual = f.readline().strip()

                q_scores = " ".join(map(str, phred33_to_q(qual)))
                writer.writerow([read_id, seq, q_scores])
            except Exception as e:
                print("Error:", e)
                break

    print("저장 완료:", outfile)
```

```python
import zipfile
from pathlib import Path

# CSV 파일이 있는 폴더 경로
input_dir = Path("raw_data/HV")   # 현재 폴더
output_zip = input_dir / "HV.zip"   # 압축 결과 파일명

# CSV 파일 수집
csv_files = list(input_dir.glob("*.csv"))

# 압축하기
with zipfile.ZipFile(output_zip, "w", compression=zipfile.ZIP_DEFLATED) as zipf:
    for csv_file in csv_files:
        zipf.write(csv_file, arcname=csv_file.name)  # arcname → 내부에 파일명만 저장
        print("추가됨:", csv_file.name)

print("압축 완료:", output_zip)
```
